```python
#Importing the required libraries
import pandas as pd

# Define a function to read the CSV file and handle the error
def read_credits_csv(file_path):
    try:
        # Attempt to read the CSV file normally
        credits = pd.read_csv(file_path)
    except pd.errors.ParserError as e:
        # If a ParserError occurs, print the error and the problematic line
        print(f"Error encountered: {e}")
        print(f"The error is on line: {e.lineno}")

        # Attempt to read the CSV file with error handling
        credits = pd.read_csv(
            file_path,
            # Skip the problematic line
            skiprows=[e.lineno - 1],
            # Specify the quote character to avoid issues with quotechar inside fields
            quoting=3,
            # Specify the encoding if it's not UTF-8
            encoding="utf-8",
            # Handle encoding errors by replacing problematic characters
            encoding_errors="replace",
        )
    return credits

# Call the function to read the CSV file
credits = read_credits_csv('tmdb_5000_credits.csv')
```

##Text Vectorization

```python
from sklearn.feature_extraction.text import CountVectorizer

# Sample text data
documents = [
    "I love data science and machine learning.",
    "Data science is amazing!",
    "Machine learning is a part of data science."
]

# Initialize CountVectorizer
cv = CountVectorizer(max_features=10, stop_words='english')

# Transform text data into a document-term matrix
X = cv.fit_transform(documents)

# Convert to array for better readability
print(X.toarray())

# Get feature names (words in the vocabulary)
print(cv.get_feature_names_out())
```

## Stemming

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
  y =[]

  for i in text.split():
    y.append(ps.stem(i))
  return " ".join(y)


```

```python
stem('in the 22nd century, a paraplegic marine is dispatched to the moon pandora on a unique mission, but becomes torn between following orders and protecting an alien civilization. action adventure fantasy sciencefiction cultureclash future spacewar spacecolony society spacetravel futuristic romance space alien tribe alienplanet cgi marine soldier battle loveaffair antiwar powerrelations mindandsoul 3d samworthington zoesaldana sigourneyweaver jamescameron')
```

```python
def recommend(movie):
  movie_index = new_df[new_df['title'] == movie].index[0]
  distances = similarity[movie_index]
  movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]

  for i in movies_list:
    print(new_df.iloc[i[0]].title)

```

```python
recommend('Avatar')
```
